### Explanation 1
This cell imports the libraries required for the experiment, such as TensorFlow, Keras, layers, and dataset helpers. It is the initialization step that makes the later model-building and training code possible.


```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist

# physical_devices = tf.config.list_physical_devices("GPU")
# tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

### Explanation 2
This cell loads the MNIST digit dataset. It provides grayscale handwritten digit images and their labels, which are commonly used for introductory classification experiments.


```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(-1, 28 * 28).astype("float32") / 255.0
x_test = x_test.reshape(-1, 28 * 28).astype("float32") / 255.0

# normalization and flattening cuz our model requires
# reshap(number of samples, new shape), -1 auto calculates the number of samples
# mnist gives 28x28 pixel images
```

### Explanation 3
This cell defines a feedforward neural network using the Sequential API. Each layer is stacked in order, making this the most direct way to build a simple Keras model.


```python
# Sequential API (Very convenient, not very flexible)
model = keras.Sequential(
    [
        keras.Input(shape=(28 * 28,1)),
        layers.Dense(512, activation="relu"),
        layers.Dense(256, activation="relu"),
        layers.Dense(10),
    ]
)
```

### Explanation 4
This cell defines a feedforward neural network using the Sequential API. Each layer is stacked in order, making this the most direct way to build a simple Keras model.



```python
#Alternate way for Sequential API
model = keras.Sequential()
model.add(keras.Input(shape=(784,)))
model.add(layers.Dense(512, activation="relu"))
model.add(layers.Dense(256, activation="relu", name="my_layer"))
model.add(layers.Dense(10))
```

### Explanation 5
This cell builds a neural network using the Functional API. It is useful when we want more control over the architecture than a plain stack of layers can provide.


```python
# Functional API (A bit more flexible)
inputs = keras.Input(shape=(784,))
x = layers.Dense(512, activation="relu", name="first_layer")(inputs)
x = layers.Dense(256, activation="relu", name="second_layer")(x)
outputs = layers.Dense(10, activation="softmax")(x)
model = keras.Model(inputs=inputs, outputs=outputs)
```

### Explanation 6
This cell prints the model summary. It lists the layers, output shapes, and parameter counts, which helps explain the architecture during a viva or report discussion.


```python
model.summary()
```

### Explanation 7
This cell defines a custom loss and a custom metric, then compiles the model using them. In a viva, you can explain that Keras allows user-defined training objectives when built-in options are not sufficient.


```python
from tensorflow.keras import backend as K
def custom_loss(y_true, y_pred):
    return K.mean(K.square(y_pred - y_true))
# backend or K class perform low level operations such as, mean the whole tensor
# this is mean squared error 

def custom_metric(y_true, y_pred):
    return K.mean(K.abs(y_pred - y_true))

# ṭhis is for just 
model.build(input_shape=(None, 784))
model.compile(optimizer='adam',
              loss=custom_loss,
              metrics=[custom_metric])
```

### Explanation 8
This cell converts integer class labels into one-hot encoded vectors. One-hot encoding is commonly used when the output layer predicts a probability distribution over multiple classes.


```python
y_train_one_hot = tf.keras.utils.to_categorical(y_train, num_classes=10)
model.fit(x_train, y_train_one_hot, batch_size=32, epochs=5, verbose=2)
```

### Explanation 9
This cell prints the model summary. It lists the layers, output shapes, and parameter counts, which helps explain the architecture during a viva or report discussion.


```python
model.summary()
```

### Explanation 10
This cell converts integer class labels into one-hot encoded vectors. One-hot encoding is commonly used when the output layer predicts a probability distribution over multiple classes.


```python
y_test_one_hot = tf.keras.utils.to_categorical(y_test, num_classes=10)
model.evaluate(x_test, y_test_one_hot, batch_size=32, verbose=2)
```
