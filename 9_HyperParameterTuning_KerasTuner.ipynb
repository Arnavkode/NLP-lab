### Explanation 3
This cell imports the libraries required for the experiment, such as TensorFlow, Keras, layers, and dataset helpers. It is the initialization step that makes the later model-building and training code possible.


```python
from tensorflow import keras
from keras_tuner import RandomSearch
```

### Explanation 4
This cell loads the MNIST digit dataset. It provides grayscale handwritten digit images and their labels, which are commonly used for introductory classification experiments.


```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

### Explanation 5
This cell preprocesses the input data by reshaping it and normalizing pixel values to the range 0 to 1. Normalization improves numerical stability and usually helps gradient-based training converge better.


```python
x_train=x_train.reshape(-1,28*28)/255.0
x_test=x_test.reshape(-1,28*28)/255.0
```

### Explanation 6
This cell defines a model-building function for Keras Tuner. The hyperparameter object `hp` allows the tuner to vary architecture choices such as number of units, activations, and even depth.


```python
def build_model(hp):
  model=keras.Sequential()
  model.add(keras.layers.Dense(units=hp.Int('units_1',min_value=32,max_value=512,step=128),activation=hp.Choice('act_1',['relu','tanh'])))
  model.add(keras.layers.Dense(units=hp.Int('units_2',min_value=32,max_value=512,step=128),activation=hp.Choice('act_2',['relu','tanh'])))
  model.add(keras.layers.Dense(10,activation='sigmoid'))
  model.compile(loss=keras.losses.SparseCategoricalCrossentropy(),
                optimizer=keras.optimizers.Adam(hp.Choice('learning_rate',[1e-2,1e-3,1e-4])),
                metrics=['accuracy'])
  return model
```

### Explanation 7
This cell creates a `RandomSearch` tuner. It repeatedly samples hyperparameter combinations, trains the corresponding models, and compares them using validation accuracy.


```python
tuner=RandomSearch(build_model,objective='val_accuracy',max_trials=5,directory='my_dir')
```

### Explanation 8
This cell prints the hyperparameter search space. It shows exactly which model settings the tuner is allowed to modify during optimization.


```python
tuner.search_space_summary()
```

### Explanation 9
This cell starts the hyperparameter search process. For each trial, Keras Tuner builds a model, trains it, evaluates it on validation data, and records the result.


```python
tuner.search(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```

### Explanation 10
This cell summarizes the tuning results by listing the best-performing hyperparameter configurations. It helps identify which model design choices gave the highest validation accuracy.


```python
tuner.results_summary()
```
